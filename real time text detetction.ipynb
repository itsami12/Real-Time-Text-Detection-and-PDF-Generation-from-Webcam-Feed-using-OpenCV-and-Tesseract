```python
import cv2
import pytesseract
from reportlab.lib.pagesizes import letter
from reportlab.pdfgen import canvas

# Initialize the camera (0 is the default camera)
cap = cv2.VideoCapture(0)

if not cap.isOpened():
    print("Error: Could not open video stream.")
else:
    print("Video stream opened successfully.")

# Create a canvas for the PDF
pdf_filename = "detected_text.pdf"
c = canvas.Canvas(pdf_filename, pagesize=letter)
width, height = letter

# Loop to continuously get frames from the camera
while True:
    # Capture frame-by-frame
    ret, frame = cap.read()
    
    if not ret:
        print("Error: Failed to capture image")
        break

    # Convert the frame to grayscale
    gray = cv2.cvtColor(frame, cv2.COLOR_BGR2GRAY)

    # Use Tesseract to detect text in the frame
    text = pytesseract.image_to_string(gray)

    # Display the resulting frame
    cv2.imshow('Camera', frame)
    
    # Print detected text
    print("Detected text:", text)

    # Add detected text to PDF
    c.drawString(30, height - 30, text)
    c.showPage()

    # Break the loop on 'q' key press
    if cv2.waitKey(1) & 0xFF == ord('q'):
        break

# Save the PDF
c.save()

# Release the camera and close windows
cap.release()
cv2.destroyAllWindows()

print(f"Detected text has been saved to {pdf_filename}")
```
